::: importing necessary libraries :::

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
index_columns=['Country Name','Country Code','Indicator Name','Indicator Code']
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_1958257.csv',index_col=index_columns)
df
```

Data Cleaning ::

```python
df.isnull().sum()
```

```python
df.fillna(df['1960'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

Data Cleaning Done :::

Sample data to do comparision :::

```python
samplePopulationData=df.loc[['India','Afghanistan','South Africa','China'],:]
```

```python
samplePopulationData
```

Separating Index columns and Value columns for plotting ::

```python
years = samplePopulationData.columns[2:]
population = samplePopulationData.iloc[:,2:]
```

```python
years
```

```python
population
```

Plotting Bar chart countrywise ::

```python
plt.figure(figsize=(18,5))
population.T.plot(kind="bar",ax=plt.gca(),color=['blue','yellow','red','green','pink'])
plt.xlabel('Year Wise')
plt.ylabel('Population')
plt.title('Population Bar Chart Comparison With other Country')
plt.legend(title='Country')
```

```python
#Plotting the Histogram
plt.figure(figsize=(18,5))
population.T.plot(kind='hist',bins=5,alpha=1,stacked=True,ax=plt.gca(),color=['red','blue','yellow','green','pink'])
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Histogram Comparision')
plt.legend(title='Country')
```

```python
year = population.index
pValue = population.iloc[:,2].values

plt.figure(figsize=(5,5))
plt.pie(pValue,autopct='%1.1f%%',labels=year,
        wedgeprops={'edgecolor':'white','linewidth':3.0},
        explode = [0.03,0.05,0.5,0],
#         startangle=90,
#         rotatelabels=10,
        textprops={'size':'large'}
       )
plt.title("population Distribution By Year")
# plt.legend(loc="center left")
plt.show()
```
