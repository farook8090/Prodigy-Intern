**Data Visualization**

```python
gameCounts = df['Game'].value_counts() #game count of gamewise
gameCounts
```

```python
gameCountPercent = round((gameCounts/len(df)) * 100,2)
gameCountPercent
```

```python
#plotting horizontal Bar graph
plt.figure(figsize=(8,10))
gameCountPercent.plot(kind="barh",color="lightgreen")

plt.title("Game Distribution (Percent)")
plt.xlabel("Percentage")
plt.ylabel("Game")
plt.show()
```

**Sentiment Analysis**

```python
df['Sentiment'].count()
```

```python
sns.countplot(x="Sentiment",data=df,palette='viridis')
plt.show()
```

```python
sentiment_counts = df['Sentiment'].value_counts()
print("::: Sentiment Counts ::: ")
sentiment_counts
```

```python
#create a pie chart

plt.figure(figsize=(8,8))
plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    colors=['purple','orange','yellow','red']
)
plt.title("Sentiment Distribution")
plt.show()
```

```python
sns.catplot(x="Game",hue="Sentiment", kind="count",height=10,aspect=3, data=df)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```
